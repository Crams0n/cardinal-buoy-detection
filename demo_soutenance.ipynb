```python
# ── Classique ─────────────────────────────────────────────────────────────────
%run classical/Classical_method_FINAL.ipynb

# ── k-NN Distance ─────────────────────────────────────────────────────────────
%run ML/ML_distance_based_classification.ipynb
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

_X_kd = np.array([FeatureExtractor.extract(crop) for crop in X])
_sc_kd = StandardScaler().fit(_X_kd)
_clf_kd = KNeighborsClassifier(n_neighbors=3, metric='cosine').fit(_sc_kd.transform(_X_kd), y)

def predict_knn_distance(crop):
    feat = FeatureExtractor.extract(crop).reshape(1, -1)
    pred_id = _clf_kd.predict(_sc_kd.transform(feat))[0]
    confidence = float(_clf_kd.predict_proba(_sc_kd.transform(feat))[0].max())
    return direction_names[pred_id], confidence

# ── k-NN Masques ──────────────────────────────────────────────────────────────
%run ML/ML_mask_knn_FINAL.ipynb
_extractor_56 = EnhancedMaskFeatureExtractor
_scaler_56    = scaler
_X_all_56     = X_all.copy()
_y_all_56     = y_all.copy()
_ext56   = _extractor_56()
_y_grp   = np.array([0 if l in (0,2) else 1 for l in _y_all_56])
_master  = WeightedKNN(k=3, metric='cosine'); _master.fit(_X_all_56, _y_grp)
_es_mask = np.array([l in (0,2) for l in _y_all_56])
_es_clf  = WeightedKNN(k=3, metric='cosine')
_es_clf.fit(_X_all_56[_es_mask], np.array([0 if l==0 else 1 for l in _y_all_56 if l in (0,2)]))
_nw_mask = np.array([l in (1,3) for l in _y_all_56])
_nw_clf  = WeightedKNN(k=3, metric='cosine')
_nw_clf.fit(_X_all_56[_nw_mask], np.array([0 if l==1 else 1 for l in _y_all_56 if l in (1,3)]))

def predict_knn_masques(crop):
    ym, yi = _ext56.create_yellow_mask(crop)
    bm, _  = _ext56.create_black_mask(crop)
    feat   = _scaler_56.transform([np.concatenate([
        _ext56.extract_stripe_features(ym, yi),
        _ext56.extract_stripe_features(bm, yi),
        _ext56.extract_vertical_stripe_features(ym, bm)
    ])])
    grp, grp_conf = _master.predict(feat)
    if grp[0] == 0:
        sub, sub_conf = _es_clf.predict(feat)
        pred = 0 if sub[0] == 0 else 2
    else:
        sub, sub_conf = _nw_clf.predict(feat)
        pred = 1 if sub[0] == 0 else 3
    confidence = float(grp_conf[0] * sub_conf[0])
    return reverse_map[pred], confidence

# ── k-Means + SVM ─────────────────────────────────────────────────────────────
%run ML/ML_kmeans_classification.ipynb
from sklearn.cluster import KMeans
from sklearn.svm import SVC
_km = KMeans(n_clusters=4, random_state=RANDOM_STATE, n_init=10)
_km.fit(X_all.astype(np.float64))
_km_labels = _km.labels_
_km_map = {}
_km_conf = {}
for cl in range(4):
    mask = _km_labels == cl
    if np.sum(mask) > 0:
        classes, counts = np.unique(y_all[mask], return_counts=True)
        winner = int(classes[np.argmax(counts)])
        _km_map[cl] = winner
        _km_conf[cl] = float(np.max(counts) / np.sum(counts))
    else:
        _km_map[cl] = 0
        _km_conf[cl] = 0.0
_ext68  = extractor
_sc_km  = scaler

def predict_kmeans(crop):
    ym, yi = _ext68.create_yellow_mask(crop)
    bm, _  = _ext68.create_black_mask(crop)
    feat   = _sc_km.transform([np.concatenate([
        _ext68.extract_stripe_features(ym, yi),
        _ext68.extract_stripe_features(bm, yi),
        _ext68.extract_vertical_stripe_features(ym, bm)
    ])]).astype(np.float64)
    cluster = int(_km.predict(feat)[0])
    return reverse_map[_km_map[cluster]], float(_km_conf.get(cluster, 0.0))

_svm = SVC(kernel='rbf', C=10, gamma=0.1, probability=True, random_state=RANDOM_STATE)
_svm.fit(X_aug_pca, y_augmented)

def predict_svm(crop):
    ym, yi = _ext68.create_yellow_mask(crop)
    bm, _  = _ext68.create_black_mask(crop)
    feat   = pca_aug.transform(scaler_aug.transform([np.concatenate([
        _ext68.extract_stripe_features(ym, yi),
        _ext68.extract_stripe_features(bm, yi),
        _ext68.extract_vertical_stripe_features(ym, bm)
    ])]))
    proba = _svm.predict_proba(feat)[0]
    pred_id = int(_svm.predict(feat)[0])
    return reverse_map[pred_id], float(proba.max())

# ── YOLOv8 ────────────────────────────────────────────────────────────────────
import glob as _glob
_pts = [p for p in sorted(_glob.glob('DP/runs/mlflow/**/*.pt', recursive=True)) if 'best.pt' in p]
if _pts:
    from ultralytics import YOLO as _YOLO
    _yolo = _YOLO(_pts[0])

    def predict_yolo(crop):
        import tempfile, cv2 as _cv2
        tmp = tempfile.mktemp(suffix='.jpg')
        _cv2.imwrite(tmp, _cv2.cvtColor(_cv2.resize(crop, (224,224)), _cv2.COLOR_RGB2BGR))
        res  = _yolo(tmp, verbose=False)[0]
        probs = res.probs.data.cpu().numpy()
        idx = int(probs.argmax())
        pred = res.names[idx]
        for v in ['East', 'North', 'South', 'West']:
            if v.lower() in pred.lower():
                return v, float(probs[idx])
        return pred, float(probs[idx])

    print(f'✓ YOLO ({_pts[0]})')
else:
    predict_yolo = lambda crop: ('N/A', float('nan'))
    print('⚠ YOLO : aucun best.pt')


def predict_classical_hsv(crop):
    hsv = cv2.cvtColor(crop, cv2.COLOR_RGB2HSV)
    densities = compute_band_densities(hsv, CONFIG)
    scores = {
        'North': (densities['y_bot'] - densities['y_top']) + (densities['b_top'] - densities['b_bot']),
        'South': (densities['y_top'] - densities['y_bot']) + (densities['b_bot'] - densities['b_top']),
        'East':  (densities['b_top'] + densities['b_bot']) / 2 - densities['b_mid'] + densities['y_mid'] - (densities['y_top'] + densities['y_bot']) / 2,
        'West':  (densities['y_top'] + densities['y_bot']) / 2 - densities['y_mid'] + densities['b_mid'] - (densities['b_top'] + densities['b_bot']) / 2,
    }
    ranked = sorted(scores.items(), key=lambda item: item[1], reverse=True)
    best_name, best_score = ranked[0]
    second_score = ranked[1][1] if len(ranked) > 1 else best_score
    score_values = np.array(list(scores.values()), dtype=float)
    span = float(score_values.max() - score_values.min())
    confidence = 0.0 if span <= 1e-12 else float(np.clip((best_score - second_score) / span, 0.0, 1.0))
    return best_name, confidence

METHODS = [
    {'name': 'Classique HSV (règles)', 'predict': predict_classical_hsv},
    {'name': 'k-NN Distance (features globales)', 'predict': predict_knn_distance},
    {'name': 'k-NN Masques (hiérarchique)', 'predict': predict_knn_masques},
    {'name': 'k-Means (clustering)', 'predict': predict_kmeans},
    {'name': 'SVM Augmenté (PCA)', 'predict': predict_svm},
    {'name': 'YOLOv8 (best.pt)', 'predict': predict_yolo},
]

print('\nMéthodes chargées :')
for i, method in enumerate(METHODS, start=1):
    print(f"  {i}. {method['name']}")

print('\n✅ Tout chargé.')
```

```python
IMAGE_PATH = 'WEST_1.jpg'  # Remplacez par le chemin de votre image

import time
from pathlib import Path

import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

img = cv2.cvtColor(cv2.imread(IMAGE_PATH), cv2.COLOR_BGR2RGB)
x1, y1, x2, y2 = detect_buoy(img)
crop = cv2.resize(img[y1:y2, x1:x2], (128, 128))

# Prédictions et mesures par méthode
results = []
for method in METHODS:
    start = time.perf_counter()
    pred, confidence = method['predict'](crop)
    elapsed_ms = (time.perf_counter() - start) * 1000.0
    results.append({
        'Méthode': method['name'],
        'Prédiction': pred,
        'Confiance': confidence,
        'Temps (ms)': elapsed_ms,
    })

report = pd.DataFrame(results)
report['Confiance'] = report['Confiance'].map(lambda value: 'N/A' if pd.isna(value) else f'{value * 100:.1f}%')
report['Temps (ms)'] = report['Temps (ms)'].map(lambda value: f'{value:.1f}')

# Affichage de l'image et du crop
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].imshow(img)
axes[0].add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1, lw=2, ec='yellow', fc='none'))
axes[0].set_title(Path(IMAGE_PATH).name)
axes[0].axis('off')
axes[1].imshow(crop)
axes[1].set_title('Crop 128×128')
axes[1].axis('off')
plt.tight_layout()
plt.show()

print('\nRésultats par méthode pour une seule image :')
print(report.to_string(index=False))

best_row = report.loc[report['Confiance'].replace('N/A', np.nan).str.rstrip('%').astype(float).idxmax()] if (report['Confiance'] != 'N/A').any() else None
if best_row is not None:
    print(f"\nConfiance la plus élevée : {best_row['Méthode']} -> {best_row['Prédiction']} ({best_row['Confiance']})")
```
